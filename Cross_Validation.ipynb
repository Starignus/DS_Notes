# Why using Cross Validation Technique

**The Scientific method** is the most powerful tool human's developed to acquire knowledge. **It is an empirical method consisting in capturing data from what is observed and applying rigorous scepticism about what is observed, given that cognitive assumptions can distort how one interprets the observation**. The methodic steps ensure the verification by the scientific community of any conclusion drawn. The beauty of Science is that it can always change as we gather more data, or its reproducibility is questionable. Nothing is written on a stone and can be updated at any time. **A systematic method aims to ensure our bias in knowledge shrinks by setting up clear steps** that have made that even published pre-review literature could be challenged and [retracted](https://en.wikipedia.org/wiki/Retraction_in_academic_publishing). The knowledge we have acquired so far contains bias from our cultural prejudice, and a more diverse scientific community will charge and test them by being more strict in reproducibility. **So what has to do all of the above to Cross Validation?** Training a machine learning model implies following systematic steps, validating and testing the model to achieve its goal. I**t is to follow a scientific approach to ensure the model's fitness and know its weakness**. When applied to production data, we must provide quantified information in advance on any risk/bias the model hasâ€”for example, granting a credit product to individuals could be biased towards gender. 

For any supervised learning algorithm, we need to divide our data into train, validation and test (out of time) data sets to ensure our model performs well, meaning it can be generalised for unseen dataâ€“removing over or under-fitting problems. However, to do this in a single round of splits for the whole data set will not asses statistically how this model will be generalised to independent data sets. Therefore, performing a statistical analysis to get this information can be done using a resampling method that uses different portions of the data to test and train a model on various iterations. The methodology is known as [Cross-validation](https://en.wikipedia.org/wiki/Cross-validation_(statistics)).

K-folds-explained.png

In short, **cross-validation combines (averages) measures of fitness (a selected metric) in prediction to derive a more accurate estimate of model prediction performance (given a metric or objective function)**. It will give us a consistent understanding of the data spread across splits to conclude whether the model is appropriate to bring into production.

In the ML or DS stream, a cross validation technique is usde for the next purpose:

* Model selection: When we want to test the perfrance of different modeling techniques ussing the same data set.
* Parameter tuning: Once you picked up one of the modeling techniques 
* Feature selection
